```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python

#Leitura da imagem
img = cv2.imread('relogio.tif', cv2.IMREAD_GRAYSCALE)
linhas = img.shape[0]
colunas = img.shape[1]
```

```python
# Reamostragem -  necessrio refazer os valores

#Imagem original com 1250 dpi
plt.subplot(1, 4, 1)
plt.imshow(img, cmap='gray', vmin=0, vmax=255)

#-----------------------------------------------------------------------
#-----------------------------------------------------------------------

#Imagem reamostrada com 300 dpi, a proporo  4.16
plt.subplot(1, 4, 2)
reamostragem = np.zeros((round(linhas/4.16), round(colunas/4.16)))

for i in range(round((linhas/4.16))):
    for j in range(round((colunas/4.16))):
        reamostragem[i][j] = img[round(i*4.16)][round(j*4.16)]
plt.imshow(reamostragem, cmap='gray', vmin=0, vmax=255)

#-----------------------------------------------------------------------
#-----------------------------------------------------------------------

#Imagem reamostrada com 150 dpi, a proporo  8.333
plt.subplot(1, 4, 3)
reamostragem2 = np.zeros((round(linhas/8.333), round(colunas/8.333)))

for i in range(round((linhas/8.333))):
    for j in range(round((colunas/8.333))):
        reamostragem2[i][j] = img[round(i*8.333)][round(j*8.333)]
plt.imshow(reamostragem2, cmap='gray', vmin=0, vmax=255)

#-----------------------------------------------------------------------
#-----------------------------------------------------------------------

#Imagem reamostrada com 72 dpi, a proporo  17.361
plt.subplot(1, 4, 4)
reamostragem3 = np.zeros((round(linhas/17.361), round(colunas/17.361)))

for i in range(round((linhas/17.361))):
    for j in range(round((colunas/17.361))):
        reamostragem3[i][j] = img[round(i*17.361)][round(j*17.361)]
plt.imshow(reamostragem3, cmap='gray', vmin=0, vmax=255)
```

```python
#Leitura da imagem
img2 = cv2.imread('ctskull-256.tif', cv2.IMREAD_GRAYSCALE)
linhas = img2.shape[0]
colunas = img2.shape[1]
```

```python
#Requantizao


#Quantizao em 8 bits -> max = 255
plt.subplot(2, 4, 1)
plt.imshow(img2, cmap='gray', vmin = 0, vmax = 255)

#-----------------------------------------------------------------------
#-----------------------------------------------------------------------

#Quantizao em 7 bits -> max = 128

plt.subplot(2, 4, 2)
for i in range(linhas):
    for j in range(colunas):
        img2[i][j] = (img2[i][j] * 128)/256
plt.imshow(img2, cmap='gray', vmin = 0, vmax = 127)

img2 = cv2.imread('ctskull-256.tif', cv2.IMREAD_GRAYSCALE)

#-----------------------------------------------------------------------
#-----------------------------------------------------------------------

#Quantizao em 6 bits -> max = 64
plt.subplot(2, 4, 3)
for i in range(linhas):
    for j in range(colunas):
        img2[i][j] = (img2[i][j] * 64)/256
plt.imshow(img2, cmap='gray', vmin = 0, vmax = 63)

img2 = cv2.imread('ctskull-256.tif', cv2.IMREAD_GRAYSCALE)

#-----------------------------------------------------------------------
#-----------------------------------------------------------------------

#Quantizao em 5 bits -> max = 32
plt.subplot(2, 4, 4)
for i in range(linhas):
    for j in range(colunas):
        img2[i][j] = (img2[i][j] * 32)/256
plt.imshow(img2, cmap='gray', vmin = 0, vmax = 31)

img2 = cv2.imread('ctskull-256.tif', cv2.IMREAD_GRAYSCALE)

#-----------------------------------------------------------------------
#-----------------------------------------------------------------------

#Quantizao em 4 bits -> max = 16
plt.subplot(2, 4, 5)
for i in range(linhas):
    for j in range(colunas):
        img2[i][j] = (img2[i][j] * 15)/256
plt.imshow(img2, cmap='gray', vmin = 0, vmax = 15)

img2 = cv2.imread('ctskull-256.tif', cv2.IMREAD_GRAYSCALE)

#-----------------------------------------------------------------------
#-----------------------------------------------------------------------

#Quantizao em 3 bits -> max = 8
plt.subplot(2, 4, 6)
for i in range(linhas):
    for j in range(colunas):
        img2[i][j] = (img2[i][j] * 7)/256
plt.imshow(img2, cmap='gray', vmin = 0, vmax = 7)

img2 = cv2.imread('ctskull-256.tif', cv2.IMREAD_GRAYSCALE)

#-----------------------------------------------------------------------
#-----------------------------------------------------------------------

#Quantizao em 2 bits -> max = 4
plt.subplot(2, 4, 7)
for i in range(linhas):
    for j in range(colunas):
        img2[i][j] = (img2[i][j] * 3)/256
plt.imshow(img2, cmap='gray', vmin = 0, vmax = 3)

img2 = cv2.imread('ctskull-256.tif', cv2.IMREAD_GRAYSCALE)

#-----------------------------------------------------------------------
#-----------------------------------------------------------------------

#Quantizao em 1 bits -> max = 2
plt.subplot(2, 4, 8)
for i in range(linhas):
    for j in range(colunas):
        if  img2[i][j] <= 127:
            img2[i][j] = 0
        else:
           img2[i][j] = 1 
plt.imshow(img2, cmap='gray', vmin = 0, vmax = 1)

img2 = cv2.imread('ctskull-256.tif', cv2.IMREAD_GRAYSCALE)
```
